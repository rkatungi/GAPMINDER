```python
import plotly.express as px
import plotly.graph_objects as go
```

```python
mapdata = px.data.gapminder()
mapdata.head()
```

```python
# Using Plotly express as px, source the data using the command px.data.gapminder() and save it into a dataframe.
#Based on the data, map the following information with the appropriate geographical plot. Map out:
#qn1: The mean life expectancy of all the Asian countries according to the data.
#we will introduce scope='asia'

fig = px.choropleth(mapdata, 
                    locations ="iso_alpha", 
                    color ="lifeExp", 
                    hover_name ="country",  
                    color_continuous_scale = px.colors.sequential.Plasma, 
                    scope ="asia", 
                    animation_frame ="year") 
fig.update_layout(
    title_text = 'Life Expectancy of Asian Countries', 
    )
fig.show()
```

```python
#Based on the data, map the following information with the appropriate geographical plot. Map out:
#qn2: Deviation in GDP of each country in Europe and South America.
#we introduce query("continent == ['Americas','Europe']". Data in Europe and Americas will be displayed
mapdf= px.data.gapminder().query("continent == ['Europe','Americas']")
fig = px.choropleth(mapdf,  
                    locations="iso_alpha",  
                    color="gdpPercap",  
                    hover_name="country",
                    ) 
fig.update_layout(
    title_text = 'GDP of Countries in Europe and Americas', 
    
)
fig.show()
```

```python
#Based on the data, map the following information with the appropriate geographical plot. Map out:
#qn3: The change in population of each African country in the last 3 decades.
# we introduce query("year==[1977,2007]" and its reflected in animation_frame = 'year'

mapafrica= px.data.gapminder().query("year==[1977,2007]")
fig = px.choropleth(mapafrica,  
                    locations="iso_alpha",  
                    color="pop",  
                    hover_name="country", 
                    scope="africa", 
                    animation_frame = 'year',
                                        ) 
fig.update_layout(
    title_text = 'Africa population in the last 3 Decades', 
   )
fig.show()
```
